# Load and prepare data
Here we load the corresponding data. The data gets normalized to a domain between 0 and 1 for each image individually. We also set the path to the rephased images, since they are needed for the following parts in the paper.

```python
source_path = 'data/phase_diagram_56.h5'
dest_path = 'data/phase_diagram_rephased.h5'
images, data = hd.load_dataset(source_path,
                               ['micromotion_phase', 'phase', 'freq'])
mmphase = data['micromotion_phase']
images = hd.normalize_single_images(images)
images = hd.prepare_images_tensorflow(images)
```

# Load the trained models
To train the models run 3b_train_micromotion_VAE.py

```python
encoder = tf.keras.models.load_model('networks/micromotion_rephasing/IIIb_VAE_micromotion_encoder.h5')
decoder = tf.keras.models.load_model('networks/micromotion_rephasing/IIIb_VAE_micromotion_decoder.h5')
latent_dim = int(encoder.layers[-1].units / 2)
encoder.summary()
decoder.summary()
```

```python
model = CVAE(latent_dim, encoder, decoder)
```

# Transform data

Create a batched dataset to respect GPU memory. If you do not use a GPU you noramlly have free choice.

```python
dataset = tf.data.Dataset.from_tensor_slices((images, mmphase)).batch(200)
```

```python
def rephase_images(phase_shift):
    rephased_images = np.empty((0, 56, 56))
    for b_images, b_mmphase in dataset:
        b_rephased_images = model.dephase(b_images, -b_mmphase + phase_shift).numpy()
        b_rephased_images = np.squeeze(b_rephased_images)
        rephased_images = np.append(rephased_images, b_rephased_images, 0)
    return rephased_images
```

## Actual rephasing
Here the data gets rephased to micromotion phase of 0.

```python
rephased_images = np.empty((0, 56, 56))
for b_images, b_mmphase in dataset:
    b_rephased_images = model.dephase(b_images, -b_mmphase).numpy()
    b_rephased_images = np.squeeze(b_rephased_images)
    rephased_images = np.append(rephased_images, b_rephased_images, 0)
```

# Plots with center of mass
We know that the center of mass is connected to the micromotion phase.

```python
def calculate_cms(ims):
    cms = []
    for idx in range(0, ims.shape[0]):
        img = np.squeeze(ims[idx])
        cms.append(center_of_mass(img))
    cms = np.array(cms)
    azimuthal = np.arctan2(cms[:, 1]-ims.shape[1]/2, cms[:, 0]-ims.shape[2]/2)*180/np.pi
    return azimuthal
```

```python
original_azimuthal = calculate_cms(images)
rephased_azimuthal = calculate_cms(rephased_images)
```

```python
unique_mm_phases = np.unique(data['micromotion_phase'])
highlite_mm_phase = 0.0*np.pi
highlite_lower_bound = highlite_mm_phase - 0.005
highlite_upper_bound = highlite_mm_phase + 0.005
highlite_selection_map = np.where((highlite_lower_bound < data['micromotion_phase']) & (data['micromotion_phase'] < highlite_upper_bound))[0]

plt.plot(data['phase'], original_azimuthal, 'x', mec=face_colors['blue'], rasterized=True)
plt.plot(data['phase'][highlite_selection_map], original_azimuthal[highlite_selection_map], 'x', mec=edge_colors['blue'], rasterized=True)
plt.xlabel('Shaking Phase (°)')
plt.ylabel(r'$\theta_{\mathrm{CoM}}$ (°)')
plt.xticks(ticks_pi_deg)
plt.yticks(ticks_pi_deg)
plt.ylim([-190, 190])
plt.xlim([-190, 190])
plt.grid()
#plt.savefig("plots/IIIb/before_micromotion_removal.pdf", bbox_inches='tight')
plt.show()
```
